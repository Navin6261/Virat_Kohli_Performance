```python
import pandas as pd
import numpy as numpy
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Virat_Kohli_odi.csv")
```

```python
df
```

```python
df.dtypes
```

```python
df = df.drop([15])
```

```python
df
```

```python
df.dtypes
```

```python
df[["Year","Match","Innings","NotOuts","100s","50s","0s","Runs","CatchesTaken"]]=df[["Year","Match","Innings","NotOuts","100s","50s","0s","Runs","CatchesTaken"]].astype("int")
```

```python
df.rename(columns = {'100s':'century','50s':'fifties','0s':'ducks'}, inplace = True)
df
```

```python
# plotting the bar chart of centuries year wise
plt.figure(figsize=(20,8))
plt.bar(df.Year,df.century)
plt.title('Centuries Year wise',color='green',size=30)
plt.xlabel('Year',size=20)
plt.ylabel('100s',size=20)
plt.show()
```

```python
#year wise centuries, fifties and ducks.

plt.figure(figsize=(20,8))
plt.bar(df.Year,df.ducks,label='ducks')
plt.plot(df.Year,df.century,label='century')
plt.plot(df.Year,df.fifties,label='fifties')
plt.legend()
plt.title('performance',color='green',size=30)
plt.xlabel('year',size=20)
plt.ylabel('Performance',size=20)
```

```python
#YEARWISE RUNS SCORED

plt.figure(figsize=(20,8))
plt.bar(df.Year,df.Runs)
plt.title('Runs Year wise',color='blue',size=30)
plt.xlabel('Year',size=20)
plt.ylabel('Runs',size=20)
plt.show()
```

```python
#YEARWISE Average

plt.figure(figsize=(20,8))
plt.bar(df.Year,df.Average)
plt.title('Average Year wise',color='blue',size=30)
plt.xlabel('Year',size=20)
plt.ylabel('Average',size=20)
plt.show()
```

```python
#match played vs Average

plt.figure(figsize=(20,8))
plt.plot(df.Year,df.Average, label='Average')
plt.plot(df.Year,df.Match,label='match played')
plt.title('Average Match wise',color='blue',size=30)
plt.legend()
plt.xlabel('Match',size=20)
plt.ylabel('Average',size=20)
plt.show()
```

```python
#match played vs Runs scored

plt.figure(figsize=(25,8))
plt.plot(df.Year,df.Runs, label='Runs')
plt.bar(df.Year,df.Match,label='match played')
plt.title('Runs Match wise',color='blue',size=30)
plt.legend()
plt.xlabel('Match',size=20)
plt.ylabel('Runs',size=20)
plt.show()
```

```python
#match played vs Runs scored

plt.figure(figsize=(25,8))
plt.plot(df.Year,df.Runs, label='Runs')
plt.bar(df.Year,df.Average,label='Average')
plt.title('Runs and Average year wise',color='green',size=30)
plt.legend()
plt.xlabel('Runs and Average',size=20)
plt.ylabel('Runs',size=20)
plt.show()
```

```python
#YEARWISE stikerate

plt.figure(figsize=(25,8))
plt.plot(df.Year,df.Strike_Rate)
plt.title('Strike_Rate Year wise',color='blue',size=30)
plt.xlabel('Year',size=20)
plt.ylabel('Strike_Rate',size=20)
plt.show()
```
